```python
import torch
import os
from torchvision import transforms
import torchstain
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#folder containing histological images
Dataset = 'Cryonuseg'
folder_path = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/Images'
```

```python
# Function to perform Reinhard normalization
def reinhard_normalization(image):
    # Convert image to float32 for processing
    img_float = image.astype(np.float32) / 255.0

    # Split the image into individual channels (assuming RGB order)
    b, g, r = cv2.split(img_float)

    # Compute the mean and standard deviation for each channel
    means = [np.mean(channel) for channel in [r, g, b]]
    stds = [np.std(channel) for channel in [r, g, b]]

    # Define the target mean and standard deviation values
    target_mean = [0.5, 0.5, 0.5]  # Target mean for each channel
    target_std = [0.5, 0.5, 0.5]   # Target standard deviation for each channel

    # Normalize each channel using the calculated means and standard deviations
    normalized_r = (r - means[0]) * (target_std[0] / stds[0]) + target_mean[0]
    normalized_g = (g - means[1]) * (target_std[1] / stds[1]) + target_mean[1]
    normalized_b = (b - means[2]) * (target_std[2] / stds[2]) + target_mean[2]

    # Merge the normalized channels back into an image
    normalized_image = cv2.merge([normalized_b, normalized_g, normalized_r])

    # Scale the values back to 0-255 and convert to uint8
    normalized_image = (normalized_image * 255).astype(np.uint8)

    return normalized_image
```

```python
output_folder = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/Reinhard'
# Loop through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.tif') or filename.endswith('.jpg') or filename.endswith('.png'):
        # Read the image
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)

        # Apply Reinhard normalization
        normalized_image = reinhard_normalization(image)

        # Save the normalized image
        save_path = os.path.join(output_folder, f'normalized_{filename}')
        cv2.imwrite(save_path, normalized_image)

        print(f"Normalized image '{filename}' saved as 'normalized_{filename}'")

print("Normalization process completed for all images.")
```
