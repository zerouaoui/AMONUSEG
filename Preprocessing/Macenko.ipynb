```python
import torch
import os
from torchvision import transforms
import torchstain
import cv2
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
#Datasets to process
#Dataset = 'Cryonuseg'
#Dataset = 'TNBC'
Dataset = 'Monuseg'
#Dataset = 'Africa'

#Input and output folders
folder_path = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/Images'
output_folder = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/Macenko'
```

```python
#define the Macenko Normlizer using torchvision
def macenko_normalization(image):
  target = cv2.cvtColor(cv2.imread('/content/drive/My Drive/MICCAI Project/target.png'), cv2.COLOR_BGR2RGB)
  T = transforms.Compose([
      transforms.ToTensor(),
      transforms.Lambda(lambda x: x*255)
      ])
  normalizer = torchstain.normalizers.MacenkoNormalizer(backend='torch')
  normalizer.fit(T(target))
  t_to_transform = T(image)
  norm, H, E = normalizer.normalize(I=t_to_transform, stains=True)
  return norm
```

```python
# Loop through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.tif') or filename.endswith('.jpg') or filename.endswith('.png'):
        # Read the image
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)


        # Apply Macenko normalization
        normalized_image = macenko_normalization(image)
        normalized_image = np.asarray(normalized_image)

        # Save the normalized image
        save_path = os.path.join(output_folder, f'{filename}')
        cv2.imwrite(save_path, normalized_image)

        print(f"Normalized image '{filename}' saved as '{filename}'")

print("Normalization process completed for all images.")
```
