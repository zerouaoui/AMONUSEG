```python
import os
import cv2

#Datasets to process
#Dataset = 'Cryonuseg'
#Dataset = 'TNBC'
Dataset = 'Monuseg'
#Dataset = 'Africa'

# Path to the folder containing normlized images
#input_folder = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/Macenko'
#input_folder = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/Reinhard'
input_folder = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/Masks'
# Output folder for normlized augmented images
#macenko
#output_folder = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/M_Augmented'
#reinhard
#output_folder = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/R_Augmented'
#Masks
output_folder = '/content/drive/My Drive/MICCAI Project/'+Dataset+'/Masks_Augmented'
#to creat the outpur folder
os.makedirs(output_folder, exist_ok=True)
```

```python
# Augmentation transformation (for histology the best is flips)
flip_horizontal = True
flip_vertical = True

# Iterate through each image in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.tif') or filename.endswith('.png') or filename.endswith('.jpg'):
        image_path = os.path.join(input_folder, filename)

        image = cv2.imread(image_path)

        # augmentation flips
        if flip_horizontal:
            flipped_horizontal_img = cv2.flip(image, 1)  # F_horizontally
        if flip_vertical:
            flipped_vertical_img = cv2.flip(image, 0)    # F_vertically

        # Save augmented images to the output folder
        if flip_horizontal:
            cv2.imwrite(os.path.join(output_folder, f"fh_{filename}"), flipped_horizontal_img)
        if flip_vertical:
            cv2.imwrite(os.path.join(output_folder, f"fv_{filename}"), flipped_vertical_img)

print("Data augmentation completed.")
```
